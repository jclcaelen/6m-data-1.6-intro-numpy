## 3\. Boolean masks: Work with targets

```python
# **Scenario:** You have campaign response data.

names = np.array(["Ana", "Ben", "Chen", "Dana", "Eli", "Fatima", "George", "Hui"])
spend = np.array([200, 150, 300, 120, 180, 220, 160, 310])      # marketing spend
revenue = np.array([400, 180, 500, 100, 220, 260, 150, 600])   # revenue
```

```python
# 1. Compute the **ROI** for each person: `roi = revenue / spend`.  
roi = revenue / spend
roi
```

```python
# 2. Create a boolean mask for customers with `roi >= 2.0`. 
roi_mask = roi >= 2.0
roi_mask
```

```python
# 3. Use the mask to:  
#   - List their names  
#   - List their spend and revenue 
names_high_roi = names[roi_mask]
print(f"Names of High ROI: ", names_high_roi)
spend_high_roi = spend[roi_mask]
print(f"Spending with High ROI:", spend_high_roi)
revenue_high_roi = revenue[roi_mask]
print(f"Revenue of High ROI: ", revenue_high_roi)
```

```python
# 4. Create a second mask for customers with `spend >= 200`.  
spending_mask = spend >= 200
```

```python
# 5. Combine the two masks to find customers who have **roi \>= 2.0 AND spend \>= 200**. 
names_high_roi_and_spend = names[roi_mask & spending_mask]
print(f"Names of High ROI and Spend: ", names_high_roi_and_spend)
```

```python
# Additional suggested solution, to do a combined mask
combined_mask = roi_mask & spending_mask
names_combined = names[combined_mask]
print(f"Names of customers with ROI >= 2.0 AND Spend >= 200: ", names_combined)
```

## 4\. Gradebook: Aggregations and broadcasting

```python
# Reuse the “Gradebook” idea from class and extend it.

students = np.array(["S1", "S2", "S3", "S4", "S5"])
subjects = np.array(["Math", "Stats", "Python"])

scores = np.array([
    [75, 80, 85],
    [60, 65, 70],
    [90, 88, 92],
    [82, 79, 84],
    [70, 72, 78]
])
```

```python
# 1. Compute each student’s **average score** (per row).  
average_scores = np.mean(scores, axis=1)  # axis=1 computes the mean across columns (subjects) for each row (student)
average_scores
```

```python
# 2. Compute each subject’s **average score** (per column).  
average_scores_subjects = np.mean(scores, axis=0)
average_scores_subjects
```

```python
# 3. Create `scores_centered` by subtracting the **subject mean** from each column (broadcasting).  
scores_centered = scores - average_scores_subjects  # This will broadcast the average scores of each subject across the respective columns
scores_centered
```

```python
# 4. For `scores_centered`, compute per-student averages again.  
scores_centered_avg = np.mean(scores_centered, axis=1)
scores_centered_avg
```

## 5\. Reshape and flatten: From daily to weekly

```python
# **Scenario:** You have 30 days of daily website visits, and you want to summarize them by week.

daily_visits = np.array([
    120, 130, 125, 140, 150,
    160, 170, 155, 145, 135,
    128, 132, 138, 142, 148,
    152, 158, 162, 168, 172,
    180, 190, 185, 175, 165,
    155, 145, 135, 125, 115
])
```

```python
daily_visits.shape
```

```python
# 1. Reshape `daily_visits` into a `(6, 5)` array representing **6 weeks × 5 days**.  
weekly_visits = daily_visits.reshape(6, 5)  # Reshape the 1D array into a 2D array with 6 rows and 5 columns
weekly_visits
```

```python
weekly_visits.shape
```

```python
# 2. Compute **total visits per week**.  
total_visits_per_week = np.sum(weekly_visits, axis=1)  # Sum across columns (days) for each row (week)
total_visits_per_week
```

```python
# 3. Compute **average visits per day of the week** (e.g., all “day 1 of week” together, all “day 2 of week” together, etc.).  
average_visits_per_day = np.mean(weekly_visits, axis=0)  # Mean across rows (weeks) for each column (day)
average_visits_per_day
```

```python
# 4. Flatten the reshaped array back to 1D and confirm it matches the original `daily_visits`.
weekly_visits_flattened = weekly_visits.flatten()  # Flatten the 2D array back to 1D
weekly_visits_flattened
```

```python
# - Generate 1,000 random test scores with `np.random.randn`, scale them to have mean 70 and standard deviation 10, then:  

test_scores = np.random.randn(1000) * 10 + 70  # Scale to mean 70 and std dev 10. Why 10 and 70? Because the standard normal distribution has a mean of 0 and a standard deviation of 1, so multiplying by 10 scales the standard deviation to 10, and adding 70 shifts the mean to 70.
test_scores[:10]  # Show the first 10 test scores to confirm they look reasonable
```

```python
# In anticipation of the next step, to check if there are any values that are outside the expected range of 0 to 100, we can do:
# Check for any scores outside the range of 0 to 100
out_of_range_scores = test_scores[(test_scores < 0) | (test_scores > 100)]
print(f"Number of out-of-range scores: {len(out_of_range_scores)}")
```

```python
#  - Clip scores to the range `[0, 100]` 
test_scores_clipped = np.clip(test_scores, 0, 100)  # This will set any values below 0 to 0 and any values above 100 to 100
```

```python
#  - Compute min, max, mean, and standard deviation  
print(f"Min: {test_scores_clipped.min()}")
print(f"Max: {test_scores_clipped.max()}")
print(f"Mean: {test_scores_clipped.mean()}")
print(f"Standard Deviation: {test_scores_clipped.std()}")
```

```python
"""
- Simulate a small A/B test:  
  - Two groups of 20 users each  
  - Randomly generate conversions (0 or 1\) for each group  
  - Compute conversion rate per group using pure NumPy operations
"""

# Simulate conversions for Group A and Group B
np.random.seed(42)  # Set a seed for reproducibility
group_A_conversions = np.random.choice([0, 1], size=20)  # Randomly generate 0s and 1s for Group A

# .choice() is a NumPy function that generates a random sample from a given 1-D array. In this case, we are using it to randomly select either 0 or 1 for each of the 20 users in Group A, simulating whether each user converted (1) or did not convert (0).
group_B_conversions = np.random.choice([0, 1], size=20)  # Randomly generate 0s and 1s for Group B

# Compute conversion rates
conversion_rate_A = group_A_conversions.mean()  # The mean of the binary array gives the conversion rate
conversion_rate_B = group_B_conversions.mean()  # The mean of the binary array gives the conversion rate
print(f"Conversion Rate for Group A: {conversion_rate_A:.2f}")
print(f"Conversion Rate for Group B: {conversion_rate_B:.2f}")
```
